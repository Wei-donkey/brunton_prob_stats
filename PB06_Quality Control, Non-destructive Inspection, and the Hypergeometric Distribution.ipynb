### "Probability Bootcamp" 6: Quality Control, Non-destructive Inspection, and the Hypergeometric Distribution
In this lecture, Dr. Brunton introduces a relevant example of the multinomial distribution: quality control and non-destructive inspection. If we test a sub-sample of size "r" and measure "m" defective items, what can we say about the reliability of the overall batch?.   

https://www.youtube.com/watch?v=e7RAK_iQBp0&list=PLMrJAkhIeNNR3sNYvfgiKgcStwuPSts9V&index=6

```python
import matplotlib.pyplot as plt
import math

def hypergeometric_prob(N, G, n, g):
    """
    Calculate the probability of getting exactly g good samples.
    """
    B = N - G  # Total defective items
    b = n - g  # Defective samples in selection
    
    # Validate constraints
    if b < 0 or b > B or g > G:
        return 0.0
    
    # Formula: [C(G, g) * C(B, b)] / C(N, n)
    numerator = math.comb(G, g) * math.comb(B, b)
    denominator = math.comb(N, n)
    
    return numerator / denominator

# 1. Define Parameters
N = 100   # Total population size
G = 80  # Total number of good items in population
n = 40   # Number of samples drawn

# 2. Calculate Probabilities for all possible values of g
# g ranges from max(0, n - B) to min(n, G)
min_g = max(0, n - (N - G))
max_g = min(n, G)
g_values = list(range(min_g, max_g + 1))
probabilities = [hypergeometric_prob(N, G, n, g) for g in g_values]

# 3. Plot the Distribution
plt.figure(figsize=(10, 6))
plt.bar(g_values, probabilities, color='skyblue', edgecolor='black')
plt.xlabel('Number of Good Samples (g)')
plt.ylabel('Probability')
plt.title(f'Hypergeometric Distribution (N={N}, G={G}, n={n})')
plt.xticks(g_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

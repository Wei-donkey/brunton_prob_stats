### "Statistics and Data Analysis" 2: Population Statistics and Random Sampling
In this lecture, Dr. Brunton introduces the notion of random sampling of a population to determine its statistics. For example, we may estimate the population mean and variance using the sample mean and variance, and we can also quantify the uncertainty in these estimates, depending on the sample size.

https://www.youtube.com/watch?v=OlkL1YatyHI&list=PLMrJAkhIeNNT14qn1c5qdL29A1UaHamjx&index=2

```python
import numpy as np
# Reproducibility (optional)
#np.random.seed(42)

# 1. Generate a large population (Poisson distribution with Lambda=10)
population_size = 20000
lambda_value = 10
data_population = np.random.poisson(lam=lambda_value, size=population_size)

# 2. Draw a smaller random sample from the population without replacement
sample_size = 200
data_sample = np.random.choice(data_population, sample_size, replace=False)

# 3. Compute Statistics of both the population and sample
pop_mean = np.mean(data_population)
pop_std = np.std(data_population) # Note: Dr. Brunton uses 1/N for population

sample_mean = np.mean(data_sample)
sample_std = np.std(data_sample) # Note: Dr. Brunton uses 1/n for sample variance here

print(f"Population Mean: {pop_mean:.4f}, Std: {pop_std:.4f}")
print(f"Sample Mean: {sample_mean:.4f}, Std: {sample_std:.4f}")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 4. Plotting
plt.figure(figsize=(10, 6))
# generate a histogram of the population data
counts, bins, _ = plt.hist(data_population, bins=19, density=True, alpha=0.6, label='Population Histogram')

# plot the poisson distribution that best fits the sample mean
x = np.arange(0, bins[-1] + 1)
pmf_values = poisson.pmf(x, mu=sample_mean)
plt.plot(x, pmf_values, 'k-', marker='o', lw=2, label=f'Poisson PMF (mean={pop_mean:.2f})')

# label the axes
plt.xlabel('Number of Events')
plt.ylabel('Probability Density')
plt.title("Population Data vs. Sample Estimate")

plt.legend()
plt.show()
```

```python
# np.random.seed(42)
# Parameters for resampling
num_samples = 1000
n = 30 # Smaller sample size
# initialize arrarys to store sample means and standard deviations
sample_means = np.zeros(num_samples)
sample_stds = np.zeros(num_samples)

# Repeat the sampling process 100 times
for i in range(num_samples):
    resample = np.random.choice(data_population, n)
    sample_means[i] = np.mean(resample)
    sample_stds[i] = np.std(resample)

# Plotting the distribution of sample means
plt.figure(figsize=(10,6))
plt.hist(sample_means, bins=15, density=True, alpha=0.6, color='blue', label='Distribution of x-bar')

# Normal Approximation (Central Limit Theorem)
from scipy.stats import norm
mu = pop_mean
# Standard Error = pop_std / sqrt(n)
std_error = pop_std / np.sqrt(n)

x_axis = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x_axis, norm.pdf(x_axis, mu, std_error), 'y-', lw=3, label='CLT Approximation')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.title(f"Distribution of Sample Means (n={n})")
plt.show()
```

### "Probability Bootcamp" 17: Parameter Estimation and Fitting Distributions
In this lecture, Dr. Brunton introduces the Poisson distribution which is one of the most important probability distributions, quantifying the probability of rare events occuring.  It is another limit of the binomial distribution, but for low-probability events.    

https://www.youtube.com/watch?v=7vXLH2H6fZw&list=PLMrJAkhIeNNR3sNYvfgiKgcStwuPSts9V&index=17

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

# Parameters from Dr. Brunton's "lights out" example
n_final = 10000
p_final = 0.002
lam = n_final * p_final # lambda = 20

# Compare Binomial distributions to the Poisson limit
n_list = [50, 200, 1000, 10000] # Increasing n to show convergence
x = np.arange(0, 50)     # Range of k (number of lights going out)

plt.figure(figsize=(12, 7))

# Plot the Poisson distribution (the theoretical limit)
plt.plot(x, poisson.pmf(x, lam), 'ko-', label=f'Poisson ($\\lambda={lam}$)', alpha=0.8, linewidth=2)

# Plot Binomial distributions to show they approach the Poisson curve
for n in n_list:
    p = lam / n
    plt.step(x, binom.pmf(x, n, p), label=f'Binomial (n={n}, p={p:.3f})', where='mid')


plt.title(f'Convergence of Binomial to Poisson ($\\lambda={lam}$)')
plt.xlabel('Number of events (k)')
plt.ylabel('Probability P(X=k)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python

plt.figure(figsize=(12, 6))

# Plot the Poisson distribution (the theoretical limit)
plt.plot(x, poisson.pmf(x, lam), 'ko-', label=f'Poisson ($\\lambda={lam}$)', alpha=0.8, linewidth=2)

# Highlight specific probabilities for k = 1, 5, 20 for lambda = 20
k_targets = [1, 10, 20]
for k in k_targets:
    prob_k = poisson.pmf(k, lam)
    plt.scatter(k, prob_k, color='red', zorder=5, s=80)
    plt.annotate(f'k={k}\nP={prob_k:.4e}', (k, prob_k), textcoords="offset points", 
                 xytext=(12,5), ha='left', fontsize=10, color='red', weight='bold')

plt.title(f'Probabilities of k lights go out ($\\lambda={lam}$)')
plt.xlabel('Number of events (k)')
plt.ylabel('Probability P(X=k)')
plt.ylim(-0.01, 0.1)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 3. Print the calculated probabilities
print(f"Poisson Probabilities for lambda = {lam}:")
for k in k_targets:
    print(f"P(X={k}) = {poisson.pmf(k, lam):.8f}")
```

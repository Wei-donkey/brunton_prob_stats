### "Statistics and Data Analysis" 14: P-hacking
In this lecture, Dr. Brunton shows how to misuse hypothesis testing, accidentally or intentionally, by p-hacking and other forms of bad statistics.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binomtest
```

```python

# Set the random seed
# Seed 41 was shown to dip below 0.05 at n=33
# Seed 42 typically stays above the threshold
np.random.seed(41)

# Parameters
n_initial = 20
n_max = 50
p_fair = 0.5  # Hypothesis: The coin is fair

# Generate an initial sequence of 20 fair coin flips (0 for tails, 1 for heads)
flips = list(np.random.binomial(1, p_fair, n_initial))

n_values = []
p_values = []

# Iteratively add one flip at a time and calculate the p-value
for n in range(n_initial, n_max + 1):
    if n > n_initial:
        # Add a new single flip to the existing sequence
        new_flip = np.random.binomial(1, p_fair)
        flips.append(new_flip)
    
    # Count total heads
    heads = sum(flips)
    
    # Perform a binomial test
    # Null Hypothesis: The probability of heads is 0.5
    result = binomtest(heads, n, p=p_fair)
    
    n_values.append(n)
    p_values.append(result.pvalue)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(n_values, p_values, marker='o', linestyle='-', color='b', label='P-value')
plt.axhline(y=0.05, color='r', linestyle='--', label='Significance Threshold (0.05)')

plt.title('P-Value vs. Number of Coin Flips (P-Hacking Demo)')
plt.xlabel('Number of Flips (n)')
plt.ylabel('P-value')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

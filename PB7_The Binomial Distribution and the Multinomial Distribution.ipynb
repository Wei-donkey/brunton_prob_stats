In this lecture, Dr. Brunton introduces the binomial distribution, which is one of the most important distributions in all of probability. It will be the basis of the Normal distribution and the Poisson distribution. The binomial distribution quantifies the probability of getting "r" successes in "n" Bernoulli trials. For example, the probability of getting 5 heads in 10 coin flips is Binomially distributed. He also introduces the more general multinomial distribution.

```python
import math
import matplotlib.pyplot as plt

def binomial_coefficient(n, k):
    """Calculate binomial coefficient C(n, k)."""
    return math.comb(n, k)

# def plot_binomial_coefficients(max_n=10):
"""Plot binomial coefficients for (x + y)^n for n from 0 to max_n."""
plt.figure(figsize=(12, 8))
max_n=10
n = 100
# for n in range(max_n + 1):
coefficients = [binomial_coefficient(n, k) for k in range(n + 1)]
total_counting = 2 ** n
probability = [coeff / total_counting for coeff in coefficients]

ks = list(range(n + 1))
# plt.scatter(ks, coefficients, s=1, alpha=0.6, label=f'n={n}' if n <= 5 else "")  # Only label first few
plt.bar(
    ks,  # Offset x-values for each n
    probability,
    width=0.8,
    label=f'n={n}',
    align='center'
)

plt.title("Binomial distribution for $x$ (n = 100)")
plt.xlabel("k")
plt.ylabel("Probability")
plt.grid(True)
# plt.legend()
plt.show()
```

Multinomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multinomial

# 1. Define Parameters
n = 10 # Total number of trials
p = [0.5, 0.3, 0.2]  # Probabilities for each category (must sum to 1)
# Example: Category A (50%), Category B (30%), Category C (20%)

# 2. Generate All Possible Outcomes
# For 3 categories, we iterate over x1 and x2, then x3 = n - x1 - x2
outcomes = []
probabilities = []

for x1 in range(n + 1):
    for x2 in range(n - x1 + 1):
        x3 = n - x1 - x2
        outcome = [x1, x2, x3]
        prob = multinomial.pmf(outcome, n, p)
        outcomes.append(outcome)
        probabilities.append(prob)

# 3. Visualization Option 1: Top 15 Most Probable Outcomes
print("Top 15 Most Probable Outcomes:")
sorted_idx = np.argsort(probabilities)[::-1][:15]

plt.figure(figsize=(12, 6))
top_outcomes = [outcomes[i] for i in sorted_idx]
top_probs = [probabilities[i] for i in sorted_idx]
labels = [f'({o[0]},{o[1]},{o[2]})' for o in top_outcomes]

plt.bar(range(len(top_probs)), top_probs, color='steelblue', edgecolor='black')
plt.xticks(range(len(top_probs)), labels, rotation=45, ha='right')
plt.xlabel('Outcome (Category A, B, C)')
plt.ylabel('Probability')
plt.title(f'Multinomial Distribution (n={n}, p={p})')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 4. Visualization Option 2: Heatmap for 2 Categories (3rd is dependent)
plt.figure(figsize=(8, 6))
x1_values = range(n + 1)
x2_values = range(n + 1)
heatmap_data = np.zeros((n + 1, n + 1))

for x1 in x1_values:
    for x2 in x2_values:
        if x1 + x2 <= n:
            x3 = n - x1 - x2
            heatmap_data[x1, x2] = multinomial.pmf([x1, x2, x3], n, p)

plt.imshow(heatmap_data.T, origin='lower', cmap='YlOrRd', extent=[0, n, 0, n])
plt.colorbar(label='Probability')
plt.xlabel('Category A Count (x1)')
plt.ylabel('Category B Count (x2)')
plt.title(f'Multinomial Heatmap (n={n}, p={p})\nCategory C = {n} - x1 - x2')
plt.grid(False)
plt.tight_layout()
plt.show()

# 5. Verification: Sum of All Probabilities Should Equal 1
print(f"\nSum of all probabilities: {sum(probabilities):.6f}")
```

There are 2n people. How many possible pairs can you pair those people off?

```python
import math

def count_pairings(n):
    """Calculate the number of ways to pair off 2n people."""
    return math.factorial(2 * n) // (2 ** n * math.factorial(n))

# Loop over n and print results
print("n\tNumber of Pairings")
print("----------------------")
for n in range(1, 11):  # Example: n from 1 to 10
    pairings = count_pairings(n)
    print(f"{n}\t{pairings}")
```

### "Statistics and Data Analysis" 6: Normal Approximation to Sample Mean
In this lecture, Dr. Brunton shows that the sample mean is a normally distributed random variable, leveraging ideas from the Central Limit Theorem (CLT). This will help us quantify the error of our sample mean approximation, and say things about the larger population.

https://www.youtube.com/watch?v=Arbj9SoU9Cs&list=PLMrJAkhIeNNT14qn1c5qdL29A1UaHamjx&index=6

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Reproducibility (optional)
#np.random.seed(42)

# Generate a large population (Poisson distribution with Lambda=10)
lambda_value = 10
N = 10000
data_population = np.random.poisson(lam=lambda_value, size=N)

# Parameters for resampling
n = 30 # sample size
m = 100 # number of samples to take

# initialize arrarys to store sample means and standard deviations
sample_means = np.zeros(m)
sample_stds = np.zeros(m)

# Repeat the sampling process m times
for i in range(m):
    resample = np.random.choice(data_population, size=n, replace=False)
    sample_means[i] = np.mean(resample)
    sample_stds[i] = np.std(resample)

# Compute the population mean and standard deviation
mean_pop = np.mean(data_population)
std_pop = np.std(data_population) # Note: Dr. Brunton uses 1/N for population

# Compute the standard error of the sample mean
std_error = std_pop / np.sqrt(n)

# Generate the normal distribution that approximates the sample mean
x = np.linspace(min(sample_means), max(sample_means), 100)
normal_approx = norm.pdf(x, loc=mean_pop, scale=std_error)

# Plotting the histogram of the sample means
plt.figure(figsize=(10,6))
plt.hist(sample_means, bins=15, density=True, alpha=0.6, color='red', edgecolor='black', label='Sample Means')

# plotting the normal approximation on top of the histogram
plt.plot(x, normal_approx, 'y-', lw=3, label=f'Normal Approximation\n mean={mean_pop:.2f}, std={std_error:.2f}')

# label the axes
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Histogram of Sample Means with Normal Approximation')

plt.legend()

plt.show()
```

### "Statistics and Data Analysis" 25: The Student's t-distribution in Statistics
In this lecture, Dr. Brunton introduces the Student's t-distribution that is an important distribution in statistics for hypothesis testing with small sample size n.   

https://www.youtube.com/watch?v=kQoPUR0hQNo&list=PLMrJAkhIeNNT14qn1c5qdL29A1UaHamjx&index=25

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Range of x values for plotting
x = np.linspace(-5, 5, 1000)

# Degrees of freedom to demonstrate convergence
degree_of_freedom = [1, 2, 5, 10, 30, 100]

#Use the inerted viridis colormap (viridis_r for reverse viridis)
cmap = plt.get_cmap('viridis_r', len(degree_of_freedom))

# Set the background color to white
plt.figure(figsize=(10,6), facecolor='white')
ax = plt.gca()
ax.set_facecolor('white')


# plot the normal distribution (standard normal, mean=0, std=1)
plt.plot(x, norm.pdf(x), 'k--', lw=2, label='Stadard Normal Distribution')

# Plot the t-distributions for different degrees of freedom, using the inverted colormap
for i, df in enumerate(degree_of_freedom):
    color = cmap(i)
    plt.plot(x, t.pdf(x, df), color=color, lw=2, label=f't-distribution (df={df})')
    
# Adding labels and title
plt.title('Convergence of t-Distribution to Normal Distribution', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('PDF', fontsize=12)
plt.legend()
plt.grid(True)

# Adjust tick color for the white color scheme
plt.xticks(color='black')
plt.show()
```

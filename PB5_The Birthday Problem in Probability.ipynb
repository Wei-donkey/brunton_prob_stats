### "Probability Bootcamp" 5: The Birthday Problem in Probability
In this lecture, Dr. Brunton introduces the birthday problem that is fun and surprisingly challenging: How many people need to be in a room before the probability that two people share the same birthday is at least 50%?  To solve this problem, we find that it is much easier to compute 1 minus the probability that two people don't share a birthday.  This idea will be very useful in other problems.   

https://www.youtube.com/watch?v=m7hI2LulMxE&list=PLMrJAkhIeNNR3sNYvfgiKgcStwuPSts9V&index=5

```python
prob = 1
days = 365
n = 50
for i in range(1,n+1):
    prob *= (days-i)/days
    print(f'number of people {i+1}, possibility of no one share birthday {round(100*  prob,2)}')
```

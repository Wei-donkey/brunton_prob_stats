### "Probability Bootcamp" 15: The Normal Distribution- The Limit of Binomial Distribution for Large n
In this lecture, Dr. Brunton introduces the Normal distribution, or "Gaussian", which is one of the most important probability distributions. It is the limit of a Binomial distribution for large "n" (for a large number of underlying Bernoulli trials). The normal distribution is quite common across statistics, for example to approximate the height of adults in a country. We will find that the sums of many random variables will often converge to a Normal distribution, according to the central limit theorem, which has broad and sweeping implications in statistics.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as spsp
from scipy.stats import norm
```

```python
n, p = 100, 0.5
q = 1-p

def binomial_pmf(k, n, p):
    return spsp.comb(n, k)* p**k * q**(n-k)

def normal_pdf(x, mu, sigma):
    return 1/(np.sqrt(2*np.pi)*sigma)*np.exp(-(x-mu)**2/2/sigma**2)

# def normal_pdf(x, mu, sigma):
#     return norm.pdf(x, loc=mu, scale=sigma)

# Generate x values for plotting
x_discrete = np.arange(0, n+1)
p_binomial = [binomial_pmf(k, n, p) for k in x_discrete]

x_continuous = np.linspace(0, n, 1000)
mu, sigma = n*p, np.sqrt(n*p*q)
p_normal = [normal_pdf(x, mu, sigma) for x in x_continuous]

plt.figure(figsize=(12,6))
plt.plot(x_discrete, p_binomial, 'b-', label='Binomial PMF')
plt.plot(x_continuous, p_normal, 'r--', label='Normal PDF')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.xlim(0,n)
```

### "Probability Bootcamp" 16: The Standard Unit Normal and Probability Computation
It is often helpful to work with a "standard unit normal" distribution, which is a Gaussian with zero mean and unit standard deviation. It is possible to transform other normal distributions into this standard form, through a simple mean-centering and scaling procedure, after which it is easy to compute probabilities.  For example, if one wants to know what percentile their height is in the USA, we could first transform the normal distribution of heights to standard unit normal, and then see where their transformed height lies. This will allow us to easily compute various quantities.

Compute the cumulative probability using CDF

```python
from scipy.stats import norm
from scipy.integrate import quad

# manually compute the CDF (using numerical integration))
mu, sigma = 200, 10  
def normal_pdf(t):
    return norm.pdf(t, mu, sigma)

# a, b = -sigma*2,sigma*2
a, b = 190, 230
# Compute cumulative probability using numerical integration
cumulative_prob, _ = quad(normal_pdf, a, b)

print(f"Standard normal cumulative probability P({a}<=Z <= {b}) = {cumulative_prob:.4f}")

# Compute cumulative probability using scipy.stats.norm.cdf
cumulative_prob_scipy = norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma)

print(f"Scipy normal cumulative probability P({a}<=Z <= {b}) = {cumulative_prob_scipy:.4f}")
```
